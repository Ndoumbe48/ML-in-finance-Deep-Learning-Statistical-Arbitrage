```python
import pickle
import numpy as np
import json

with open('results/OUFFN/results_dear-python_1777034455.pickle', 'rb') as f:
    results = pickle.load(f)

for model_name, data in results.items():
    print("\n" + "="*80)
    print(f"MODEL: {model_name}")
    print("="*80)
    
    # Display key metrics
    if isinstance(data, dict):
        for key, val in data.items():
            if key == 'config':
                print(f"\n{key}:")
                print(json.dumps(val, indent=2, default=str))
            elif key == 'returns':
                print(f"\n{key}:")
                print(f"  Shape: {np.array(val).shape}")
                print(f"  Min: {np.min(val):.6f}")
                print(f"  Max: {np.max(val):.6f}")
                print(f"  Mean: {np.mean(val):.6f}")
                print(f"  First 10: {val[:10]}")
            elif key == 'timestamp':
                print(f"\n{key}: {val}")
            else:
                print(f"\n{key}: {val}")
```
